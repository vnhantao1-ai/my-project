```python
'''  In các số dương lẻ bé hơn 100  '''
for i in range (1, 101):
    if i % 2 == 1:
        print(i, end = " ")
```

```python
'''  In các số dương chẵn chia hết cho 3 bé hơn 100 '''
for i in range (1, 101):
    if i % 2 == 0 and i % 3 == 0:
        print(i, end = " ")
```

```python
''' Nhập vào số nguyên dương a, in ra bảng cửu chương của a  '''
try:
    a = int(input())
    if a > 0:
        for i in range (1, 10):
            print(a*i, end = " ")
    else:
        print("Nhap lai")
except ValueError:
    print("Nhap lai")
```

```python
''' Bài 4: Viết chương trình in ra hình tam giác cân có độ cao h được nhập từ bàn phím '''
h = int(input())
for i in range (h):
    print(" "*(h-i-1), end = "")
    for k in range (2*i + 1):
        print("*", end = "")
    print()
```

```python
''' Bài 5: Nhập vào số nguyên dương n. Tính S = 1 + 2 + 3 + 4 + … + n '''
n = int(input())
S = 0
for i in range (n+1):
    S += i
print(S)
```

```python
'''Nhập vào số nguyên dương a, in ra tất cả các ước dương của a '''
import math as m
a = int(input())
for i in range (1, int(m.sqrt(a))+1):
    if a % i == 0:
        print(i, a//i)
    else:
        continue
```

```python
'''Bài 7: Nhập vào số nguyên dương a, đếm số ước dương của a '''
import math as m
a = int(input())
tong_uoc = 0
for i in range (1, int(m.sqrt(a))+1):
    if a % i == 0:
        tong_uoc += 1
    else:
        continue
print(tong_uoc*2)
```

```python
''' Bài 8: Nhập vào số nguyên dương a và b, in toàn bộ ước dương chung của a và b '''
import math as m
a, b = map(int, input().split())
if a > b:
    for i in range (1, int(m.sqrt(b))):
        if a % i == 0 and b % i == 0:          # Tim uoc ben trai can                                           
            print(i)                            # Tim uoc phai can
            if a % (b//i) == 0:
                print(a//i)
        else:
            continue
elif a == b:
    print(f"1, {a}")
else:
    for i in range (1, int(m.sqrt(a))):
        if a % i == 0 and b % i == 0:
            print(i)
            if b % (a//i) == 0:
                print(a//i)
        else:
            continue
```

```python
#A4 Tam giác rỗng
h = int(input())
for i in range (1, h+1):
    print(" "*(h-i), end = "")
    if i == 1:
        print("*")
    elif i == h:
        print("*"*(2*h-1))
    else:
        print("*", end = "")
        print(" "*(2*i-3), end = "")
        print("*")
```
