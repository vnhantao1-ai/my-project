```python
#W11A1
def reverse(path: str) -> int:
    with open(path, 'r') as f:
        data = f.read().strip()
        data_reversed = data[::-1]
        return int(data_reversed)
```

```python
#W11A4
def areAnagrams(path1: str, path2: str) -> bool:
    with open(path1, "r") as f1:
        str1 = f1.read()
    with open (path2, "r") as f2:
        str2 = f2.read()
    
    str1 = str1.lower()
    str2 = str2.lower()

    str1 = "".join(str1.split())
    str2 = "".join(str2.split())

    if sorted(str1) == sorted(str2):
        return True
    
    return False

with open("path1.txt", "w") as f1:
    f1.write("Tom Marvolo Riddle")
with open("path2.txt", "w") as f2:
    f2.write("I Am Lord Voldemort")

areAnagrams("path1.txt", "path2.txt")
```

```python
#W11A5
def  count(path: str, character: str) -> int:
    with open(path, "r") as f:
        str1 = f.read()
    
    result = 0
    for i in str1:
        if i == character:
            result += 1
    
    return result

with open("file.txt", "w") as f:
    f.write("We accept the love we think we deserve.")

count("file.txt", "e")
```

```python
#W11A12
def maximumProduct(path: str) -> int:
    with open(path, "r") as f:
        n = f.readline()
        line2 = f.readline()   
        nums = [int(x) for x in line2.split()]

        nums.sort()

        option1 = nums[-1] * nums[-2] * nums[-3]

        option2 = nums[0] * nums[1] * nums[-1]
        
        return max(option1, option2)
```

```python
#W11A13
def moveZeroes(fileName: str) -> list:
    with open(fileName, "r") as f:
        n_line = f.readline()
        content = f.readline()
        nums = [int(x) for x in content.split()]

        non_zeroes = [x for x in nums if x != 0]
        zeroes = [x for x in nums if x == 0]
        
        return non_zeroes + zeroes
```

```python
#W11A14
def averageTime(path: str):
    with open(path, "r", encoding="utf-8") as f:
        f.readline()
        total_time = 0.0  
        count = 0        

        while True:
            line = f.readline()

            if not line:
                break
                
            line = line.strip()
            parts = line.split(',')
            total_time += float(parts[1])

            count += 1
            
            avg = total_time / count
            return round(avg, 2)
```

```python
#W11A15
def findMovies(path: str) -> list[str]:
    valid_movies = []
    with open(path, "r") as f:
        f.readline()

        for line in f:
            line = line.strip()
            if not line:
                break

            parts = line.split(',')
            movie_name = parts[1].strip()
            description = parts[2].strip()
            rating_val = float(parts[3].strip())

            if rating_val > 8.0 and description.lower() != "boring":
                valid_movies.append(movie_name)

    if not valid_movies:
        return "All are boring!"
            
    return valid_movies
```

```python
#W11A16
import math
def function(path: str) :
    with open(path, "r") as f:
        while True:
            line1 = f.readline().lower().split()
            line2 = f.readline().split()
            space = f.readline()
            
            if not space:
                break

            if "square" in line1:
                result = int(line2[1]) * 4
                print(f"{result:.2f}")
            
            elif "rectangle" in line1:
                result = 2 * (int(line2[1]) + int(line2[3]))
                print(f"{result:.2f}")

            elif "circle" in line1:
                result = math.pi * 2 * int(line2[1])
                print(f"{result:.2f}")
```

```python
#W11A17
def  productExceptSelf (path: str) -> list[int]:
    with open(path, "r") as f:
        f.readline()
        data = f.readline().split()
    
    data = [int(x) for x in data]
    result = []

    pow_data = 1

    for i in data:
        pow_data *= i
    
    count_zero = data.count(0)
    pow_without_zero = 1
    
    for i in data:
        if i != 0:
            pow_without_zero *= i

    if count_zero > 1:
        return [0] * len(data)
    
    else:
        for i in data:
            if i == 0:
                result.append(pow_without_zero)
            else:
                part = pow_data // i
                result.append(part)
    
    return result
```

```python
#W11A18
import math

def estimatedTime(path: str):
    with open(path, "r", encoding="utf-8") as f:
        line1 = f.readline().split()
        x1 = float(line1[2])
        y1 = float(line1[4])

        line2 = f.readline().split()
        x2 = float(line2[2])
        y2 = float(line2[4])

        line3 = f.readline().split()
        v = float(line3[1])
   
        distance = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)

        time_needed = distance / v
            
        return round(time_needed, 2)
```
