```python
#W11A1
def reverse(path: str) -> int:
    with open(path, 'r') as f:
        data = f.read().strip()
        data_reversed = data[::-1]
        return int(data_reversed)

ten_file = "đm python"
with open(ten_file, "w") as f:
    f.write("123456")
result = reverse(ten_file)
print(result)
```

```python
def canWinNim(path: str) -> bool:
    try:
        with open(path, 'r') as f:
            content = f.read().strip()
            if not content:
                return False
            n = int(content)
            return n % 4 != 0
            
    except FileNotFoundError:
        print(f"Lỗi: Không tìm thấy tệp tại {path}")
        return False
    except ValueError:
        print("Lỗi: Nội dung tệp không phải là số nguyên hợp lệ")
        return False      
```

```python
def housesOfHogwarts(path: str):
    try:
        with open(path, 'r', encoding='utf-8') as f:
            first_line = f.readline()
            if not first_line:
                return

            n = int(first_line.strip())
            for _ in range(n):
                name = f.readline()
                if name:
                    print(name.strip())
                    
    except FileNotFoundError:
        print(f"Lỗi: Không tìm thấy tệp tại {path}")
    except ValueError:
        print("Lỗi: Dòng đầu tiên của tệp không phải là số nguyên hợp lệ.")
```
