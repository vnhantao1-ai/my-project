```python
#W11A1
def reverse(path: str) -> int:
    with open(path, 'r') as f:
        data = f.read().strip()
        data_reversed = data[::-1]
        return int(data_reversed)
```

```python
#W11A2
def canWinNim(path: str) -> bool:
    with open(path, 'r') as file:
        content = file.read().strip()
    n = int(content).
    return n % 4 != 0  
```

```python
#W11A3
def housesOfHogwarts(path: str):
    with open(path, 'r', encoding='utf-8') as f:
        line1 = f.readline()

        n = int(line1.strip())
        for _ in range(n):
            name = f.readline()
            print(name.strip())
```

```python
#W11A4
def areAnagrams(path1: str, path2: str) -> bool:
    with open(path1, "r") as f1:
        str1 = f1.read()
    with open (path2, "r") as f2:
        str2 = f2.read()
    
    str1 = str1.lower()
    str2 = str2.lower()

    str1 = "".join(str1.split())
    str2 = "".join(str2.split())

    if sorted(str1) == sorted(str2):
        return True
    
    return False

with open("path1.txt", "w") as f1:
    f1.write("Tom Marvolo Riddle")
with open("path2.txt", "w") as f2:
    f2.write("I Am Lord Voldemort")

areAnagrams("path1.txt", "path2.txt")
```

```python
#W11A5
def  count(path: str, character: str) -> int:
    with open(path, "r") as f:
        str1 = f.read()
    
    result = 0
    for i in str1:
        if i == character:
            result += 1
    
    return result

with open("file.txt", "w") as f:
    f.write("We accept the love we think we deserve.")

count("file.txt", "e")
```

```python
#W11A8
def function(n: int):
    if n == 0:
        return "a"
    elif n == 1:
        return "b"
    else:
        result = function(n-1) + function(n-2)
    
    return str(result)

n = int(input())
with open("output.txt", "w") as f:
    f.write(function(n))
```

```python
#W11A9
def grade10(path: str):
    with open(path, "r", encoding="utf-8") as f:
        lines = f.readlines()

        for i in range(1, len(lines)):
            line = lines[i].strip() 

            parts = line.split(',')

            diem_THCS4 = parts[5].strip()
            if diem_THCS4 == "10":
                ho = parts[1].strip()
                ten = parts[2].strip()
                print(f"{ho} {ten}")

content = """STT,Ho,Ten,Ngay sinh,Dai so,THCS4,THCS1
1,Truong Vo,Ky,10/10/1999,8,10,9
2,Chu Chi,Nhuoc,13/9/1999,10,10,10
"""
with open ("file.csv", "w") as f:
    f.write(content)

grade10("file.csv")
```

```python
#W11A10
def getMoney(path: str) -> list[int]:
    with open(path, "r") as f:
        file = f.read() 
        a = file.split()
        result = [int(x) for x in a]
    return result

def rob(hourses: list[int]) -> int :
    n = 0
    result = 0
    while n >= 0 and n < len(hourses):
        result += hourses[n]
        n += 2
    
    return result
```
