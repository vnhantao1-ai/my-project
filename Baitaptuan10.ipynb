```python
#W9A1
def merge_recursive(list1, list2):
    if list1 == []:
        return list2
    if list2 == []:
        return list1

    if list1[0] < list2[0]:
        return [list1[0]] + merge_recursive(list1[1:], list2)
    else:
        return [list2[0]] + merge_recursive(list1, list2[1:])

L1 = list(map(int, input().split()))
L2 = list(map(int, input().split()))

print(merge_recursive(L1, L2))
```

```python
#W9A2
import ast

def merge_2_lists(a, b):
    if not a: return b
    if not b: return a
    
    if int(a[0]) < int(b[0]):
        return [a[0]] + merge_2_lists(a[1:], b)
    else:
        return [b[0]] + merge_2_lists(a, b[1:])

a = ast.literal_eval(input())
l = []
k = int(input())

for i in range(len(a)):
    l = merge_2_lists(l, a[i])

print(l[-k])
```

```python
#W9A3
def min_number_by_one_swap_recursive(n):
    s = list(str(n))
    length = len(s)

    def helper(i):
        if i == length:
            return False

        min_digit = s[i]
        pos = -1

        for j in range(i + 1, length):
            if s[j] < min_digit:
                if i == 0 and s[j] == '0':
                    continue
                min_digit = s[j]
                pos = j

        if pos != -1:
            s[i], s[pos] = s[pos], s[i]
            return True

        return helper(i + 1)

    helper(0)
    return int("".join(s))
```

```python
#W9A4
def can_transform(a, b):
    def dfs(i, j):
        if j == len(b):
            return a[i:].islower()

        if i == len(a):
            return False

        if a[i].isupper():
            return a[i] == b[j] and dfs(i + 1, j + 1)

        return dfs(i + 1, j) or (
            a[i].upper() == b[j] and dfs(i + 1, j + 1)
        )

    return "YES" if dfs(0, 0) else "NO"
```

```python
#W9A5
def is_subarray(list1, list2):
    n, m = len(list1), len(list2)

    def match(i, j):
        if j == n:
            return True

        if i + j >= m or list1[j] != list2[i + j]:
            return False
        return match(i, j + 1)

    def check(i):
        if i > m - n:
            return False

        if match(i, 0):
            return True
        return check(i + 1)

    return check(0)
```

```python
#W9A6
def next_permutation(a):
    n = len(a)

    def find_i(i):
        if i < 0:
            return -1
        if a[i] < a[i + 1]:
            return i
        return find_i(i - 1)

    def find_j(j, i):
        if a[j] > a[i]:
            return j
        return find_j(j - 1, i)

    def reverse(l, r):
        if l >= r:
            return
        a[l], a[r] = a[r], a[l]
        reverse(l + 1, r - 1)

    i = find_i(n - 2)

    if i == -1:
        reverse(0, n - 1)
        return

    j = find_j(n - 1, i)

    a[i], a[j] = a[j], a[i]

    reverse(i + 1, n - 1)
```

```python
#W9A7
def canPlaceFlowers(flowerbed, k):
    n = len(flowerbed)

    def dfs(i, k):
        if k == 0:
            return True

        if i >= n:
            return False

        can_plant = (
            flowerbed[i] == 0 and
            (i == 0 or flowerbed[i - 1] == 0) and
            (i == n - 1 or flowerbed[i + 1] == 0)
        )

        if can_plant:
            flowerbed[i] = 1
            if dfs(i + 2, k - 1):
                flowerbed[i] = 0
                return True
            flowerbed[i] = 0

        return dfs(i + 1, k)

    return dfs(0, k)
```
