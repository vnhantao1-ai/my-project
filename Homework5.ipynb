```python
#W6A16
# các hàm input phải ở bên ngoài hàm def
# f"{a[idx]:>4}" với : Dấu hai chấm báo hiệu bắt đầu phần quy định định dạng (format specifier).
#>: Dấu này có nghĩa là Căn lề phải (Right align).
#4: Độ rộng tối thiểu là 4 ký tự.
def matrix(n: int, m: int, a: list):
    idx = 0
    for i in range(n):
        for j in range(m):
            print(f"{a[idx]:>4}", end="")
            idx += 1
        print()
n, m = map(int, input().split())
a = input().split()
matrix(n, m, a)
```

```python
#W6A17
def matrix(n: int, a: list):
    if len(a) != n*n:
        print("Nhap dung so luong")
        return None
    cheo_chinh = []
    cheo_phu = []
    for i in range(n):
        cheo_chinh.append(a[i*(n+1)])
        cheo_phu.append(a[(i+1)*(n-1)])
    return cheo_chinh, cheo_phu
n = int(input())
a = list(map(int, input().split()))
cc, cp = matrix(n, a)
print("Duong cheo chinh:", *cc)
print("Duong cheo phu:", *cp)
```

```python
#W6A18
def sum_col(n: int, m: int, k: int, a: list):
    if len(a) != n*m:
        print("Nhap dung so luong")
        return None
    if k < 0 or k >= m:
        print("Cot khong hop le")
        return None
    tong = 0
    for i in range(n):
        tong += int(a[i*m + k])
    return tong
n, m, k = map(int, input().split())
a = input().split()
print(sum_col(n, m, k, a))
```

```python
#W6A19
# hàm a.index() chỉ ra vị trí của nó
def max_in_list(a):
    return a.index(max(a))
a = list(map(int, input().split()))
print(max_in_list(a))
```

```python
#W6A20
def inves(a: list, k: int):
    for i in range(len(a)):
        if a[i] == k:
            return i
    return -1
a = list(map(int, input().split()))
k = int(input())
print(inves(a, k))
```

```python
#W6A23
# 1 lỗi nguy hiểm : xoá list khi đang duyệt list
def erase(a: list, x: int):
    b = []              
    for i in a:          
        if i >= x:       
            b.append(i)
    return b             
a = list(map(int, input().split()))
x = int(input())
print(erase(a, x))
```

```python
#W6A24
def push_0(a):
    zeros = []     
    others = []     
    for i in a:
        if i == 0:
            zeros.append(i)
        else:
            others.append(i)
    return zeros + others
a = list(map(int, input().split()))
print(push_0(a))
```

```python
#W6A25
def phan_tu_thieu(a: list):
    pt = []
    for i in range(10):       
        if i not in a:
            pt.append(i)
    return pt
a = list(map(int, input().split()))
print(phan_tu_thieu(a))
```

```python
#W6A26
def tong(a: list):
    if len(a) < 3:
        return 0
    sum_a = 0
    for i in range(1, len(a)-1):
        if a[i] > a[i-1] and a[i] > a[i+1]:
            sum_a += a[i]
    return sum_a
a = list(map(int, input().split()))
print(tong(a))
```
