```python
#W6A16
def matrix(n: int, m: int, a: list):
    n, m = map(int, input().split())
    a = input().split()
    location = 0
    for i in range (n):
        for j in range (m):
            if len(a[location]) <= 4:
                print(" "*(4-len(a[location])), end = "")
                print(int(a[location]), end = "")
                location += 1
        print()
matrix(n, m, a)
```

```python
#W6A17
def matrix(n: int, m: int, a: list):
    if n != m:
        print("Nhap dung so luong")
        return None
    if len(a) != m*n:
        print("Nhap dung so luong")
        return None
    l = []
    for i in range (0, m*n, n+1):
        l.append(a[i])
    return l
n, m = map(int, input().split())
a = input().split()
print(matrix(n, m, a))
```

```python
#W6A18
def sum_col(n: int, m: int, k: int, a: list):
    if len(a) != m*n:
        print("Nhap dung so luong")
        return None
    tong = 0
    for i in range (0, m*n, n):
        tong += int(a[i])
    return tong
m, n, k = map(int, input().split())
a = input().split()
print(sum_col(n, m, k, a))
```

```python
#W6A19
def max_in_list(a: list):
    m = 0
    for i in range(len(a)):
        if int(a[i]) > m:
            m = int(a[i])
    return m
a = input().split()
print(max_in_list(a))
```

```python
#W6A20
def inves(a: list, k: int):
    a = input().split()
    k = int(input())
    if str(k) in a:
        idx = 0
        for i in range (len(a)):
            if int(a[idx]) == k:
                return idx +1
            else:
                idx += 1
    else:
        return -1
print(inves(a, k))
```

```python
#W6A23
def erase(a: list, x: int):
    for i in range (len(a)):
        if int(a[0]) < x:
            a.pop(0)
    return a
a = input().split()
x = int(input())
print(erase(a, x))
```

```python
#W6A24
def push_0(a):
    b = []
    c = ""
    for i in a:
        b.append(i)
    for i in b:
        if i == "0":
            b.remove("0")
            b.append("0")
    for i in b:
        c += i
    return c
a = input()
print(push_0(a))
```
