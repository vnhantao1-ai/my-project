```python
#W6A1
def phan_tu_trung(x):
    kq = []
    for i in x:
        if i not in kq:
            kq.append(i)
    return kq

x = list(map(int, input().split()))
print(phan_tu_trung(x))
```

```python
#W6A2
# chưa xác định được xác định dạng input
def tong_tich_luy(x):
    list_tong = []
    for i in range(len(x)):
        list_tong.append(sum(x[:i+1]))
    return list_tong
x = list(map(int, input().split()))
print(tong_tich_luy(x))
```

```python
#W6A3
# Không nên dùng hàm pop trong bài này vì sẽ không tối ưu
# chưa tối ưu cho việc nhập số k quá lớn
def xoay_trai(a: tuple, k: int) -> tuple:  
    k %= len(a)
    return a[k:] + a[:k]
a = tuple(map(int, input().split()))
k = int(input())
print(xoay_trai(a, k))
```

```python
#W6A5
def so_sanh(a):
    d = {"positives": 0,"negatives": 0,"zeros": 0}
    for i in a:
        if i > 0:
            d["positives"] += 1
        elif i < 0:
            d["negatives"] += 1
        else:
            d["zeros"] += 1
    return d
a = list(map(int, input().split()))
print(so_sanh(a))
```

```python
#W6A7
# tuple co phép slicing, nhưng không có phép reverse hay xoá hay thêm
def bien_tuple(a: tuple):
    first = a[0]
    last = a[-1]
    reversed_tuple = a[::-1]
    return first, last, reversed_tuple
a = tuple(map(int, input().split()))
print(bien_tuple(a))
```

```python
#W6A8
def dem(a: list):
    d = {}
    for i in a:
        if i not in d:
            d[i] = 1
        else:
            d[i] += 1
    return d
a = input().split()
print(dem(a))
```

```python
#W6A9
# hàm item sẽ tách từng phần tửu dict thành dạng tuple còn cả dict thành dạng list
# sorted sẽ sắp xếp các phần tử đầu tiên
# hàm dict sẽ chuyển về dạng dict
def gop(a: dict, b: dict):
    for i in (b):
        if i not in a:
            a[i] = b[i]
        else:
            a[i] = a[i] + b[i]
    return a
def dang_dict(a):
    b = {}
    for i in a:
        for j in range(len(i)):
            if i[j] == ":":
                b[i[:j]] = int(i[(j+1):])
    return b
a = input().split()
b = input().split()
kq = gop(dang_dict(a), dang_dict(b))
kq = dict(sorted(kq.items()))
print(kq)
```

```python
#W6A10
# key= như là 1 giá trị mà cần máy so sánh và máy sẽ chỉ so sánh giá trị đó
def tong_bang_k(a: list, k: int):
    b = []
    for i in range(len(a)):
        for j in range(i+1, len(a)):
            if a[i] + a[j] == k:
                b.append((a[i], a[j]))
    b.sort(key=lambda x: x[0])
    return b
a = list(map(int, input().split()))
k = int(input())
print(tong_bang_k(a, k))
```

```python
#W6A12
def max_count(a: list):
    a.sort()
    most_count = 0
    most_appearance = a[0]
    i = 0
    while i < len(a):
        count = 1
        j = i + 1
        while j < len(a) and a[j] == a[i]:
            count += 1
            j += 1
        if count > most_count:
            most_count = count
            most_appearance = a[i]
        i = j
    return most_appearance
a = list(map(int, input().split()))
print(max_count(a))
```

```python
#W6A13
# câu lệnh i.split() là để tách qua dấu :
# không được dùng việc key và value cùng index vì không tối ưu
def reverse_dict(a: dict):
    d = {}
    for k, v in a.items():
        d[v] = k
    return d

def dang_dict(a):
    b = {}
    for i in a:
        key, value = i.split(":")
        b[key] = int(value)
    return b
a = input().split()
print(reverse_dict(dang_dict(a)))
```
