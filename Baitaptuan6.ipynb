```python
#W6A2
def tong_tich_luy(x):
    b = str(x)
    list_tong = []
    list_so = []
    for i in range (len(b)):
        list_so.append(int(b[i]))
        list_tong.append(sum(list_so))
    return list_tong
x = int(input())
print(tong_tich_luy(x))
```

```python
#W6A3
def xoay_trai(a: tuple, k:int)-> tuple:
    b = list (a)
    for i in range (k):
        b.append(b[0])
        b.pop(0)
    return tuple(b)
a = tuple(input())
k = int(input())
print(xoay_trai(a, k))
```

```python
#W6A5
def so_sanh(a):
    b = list(str(a))
    d = {}
    negative = 0
    positive = 0
    zero = 0
    for i in range (len(b)):
        if  int(b[i]) > 0:
            positive +=1
        elif int(b[i]) < 0:
            negative +=1
        else:
            zero += 1
    for i in range (3):
        d["positive"] = positive
        d["negative"] = negative
        d["zero"] = zero
    return d
a = int(input())
print(so_sanh(a))        
```

```python
#W6A7
def bien_tuple(a: tuple):
    b = list(a)
    b.sort(reverse = True)
    return b[0], b[len(b)-1], tuple(b)
a = tuple(input())
print(*bien_tuple(a))
```

```python
#W6A8
def dem(a: list):
    d = {}
    for i in (a):
        if i not in d:
            d[i] = 1
        else:
            d[i] += 1
    return d
a = list(input())
print(dem(a))
```

```python
#W6A9
def gop(a: dict, b: dict):
    for i in (b):
        if i not in a:
            a[i] = b[i]
        else:
            a[i] = a[i] + b[i]
    return a
def dang_dict(a):
    b = {}
    for i in a:
        for j in range(len(i)):
            if i[j] == ":":
                b[i[:j]] = int(i[(j+1):])
    return b
a = input().split()
b = input().split()
print(gop(dang_dict(a), dang_dict(b)))
```

```python
#W6A10
def tong_bang_k(a: list, k: int):
    b = []
    for i in range (len(a)):
        for j in range (i+1, len(a)):
            if int(a[i]) + int(a[j]) == k:
                b.append((a[i], a[j]))
    return b
a = input().split()
k = int(input())
print(*tong_bang_k(a, k))
```

```python
#W6A13
def reverse_dict(a: dict):
    d = {}
    set_key = list(a)
    set_value = []
    for i in (a):
        set_value.append(a[i])
    for i in range(len(a)):
        d[set_value[i]] = set_key[i]
    return d
def dang_dict(a):
    b = {}
    for i in a:
        for j in range(len(i)):
            if i[j] == ":":
                b[i[:j]] = int(i[(j+1):])
    return b
a = input().split()
print(reverse_dict(dang_dict(a)))
```

```python
#W6A14
def same(a: list, b: list):
    l = []
    for i in range (len(a)):
        for j in range (len(b)):
            if int(a[i]) == int(b[j]):
                l.append(a[i])
    return l
a = input().split()
b = input().split()
print(same(a, b))
```
