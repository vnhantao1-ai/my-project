```python
'''hập hai số từ bàn phím và kiểm tra xem chúng có phải là cặp số thân thiết không. 
Nếu có, in ra màn hình "true", ngược lại, in ra "false"'''
try:
    a, b = map(int, input().split())
    if a > 0 and b>0:
        tong_uoc_a = 0
        for i in range (1, a//2+1):
            if a % i == 0:
                tong_uoc_a += i
        if tong_uoc_a == b:
            print("true")
        else:
            print("false")
    else:
        print("Nhap 2 so lon hon 0")
except ValueError:
    print("Nhap gia tri dung")
```

```python
'''hập vào số nguyên dương n. In ra tất cả các số chính phương ≤ n có các chữ số 
đôi một khác nhau (in theo thứ tự tăng dần, cùng hàng, cách nhau bởi một dấu 
cách). Nếu không có, in ra (no number). '''
try:
    def khac_nhau(a):
        a = str(a)
        u = 0
        for i in a:  
            for j in a: 
                if i == j: 
                    u += 1 
        if u > len(a): 
            return False
        if u <= len(a):
            return True
    n = int(input())
    if n >= 0:
        for i in range (n+1):
            if i**2 <= n and khac_nhau(i**2) == True:
                print(i**2, end = " ")
            elif i**2 > n:
                break
    else:
        print("no number")
except ValueError:
    print("Nhap lai")
```

```python
#hập số nguyên dương n. Lặp lại phép toán n = n + reverse(n) cho đến khi n là số 
#đối xứng (palindrome). In ra số bước và giá trị palindrome thu được. 
#A8
def reverse(n):
    reverse = 0
    while n>0:
        reverse = reverse*10 + n%10
        n//=10
    return reverse
n = int(input())
so_buoc = 0
while n != reverse(n):
    n = n + reverse(n)
    so_buoc += 1
print(n)
print(so_buoc)
```

```python
#Nhập vào số nguyên dương n. Tìm số x trong [1, n] có độ dài dãy Collatz lớn nhất 
#và in ra x cùng độ dài dãy tương ứng (x L). Nếu có nhiều số cùng độ dài, chọn số 
#nhỏ nhất." 
#A10
n = int(input())
best_len = 1
best_x = 1
x = 1
while x != n:
    t = x
    length = 0
    while  t != 1:
        if t % 2 == 0:
            t //= 2
        else:
            t = 3*t +1
        length += 1
    if length > best_len:
        best_len = length
        best_x = x
    elif length == best_len and best_x < x:
        pass
    x += 1
print(best_x, best_len)
```
