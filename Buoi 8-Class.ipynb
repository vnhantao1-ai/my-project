```python
#W8A1
class Clinder:
    def __init__(self, r: float, h: float):
        self.r = r
        self.h = h
    def surface_area(self):
        result = 2*3.14*self.r**2 + 2*self.r*3.14*self.h
        return result
    def volume(self):
        result = 3.14*self.r**2*self.h
        return result
r, h = map(int, input().split())
a = Clinder(r, h)
print(a.surface_area())
print(a.volume())
```

```python
#W8A2
class Date:
    def __init__(self, D: int, M: int, Y: int):
        self.D = D
        self.M = M
        self.Y = Y
    def get_max_days(self):
        if (self.Y % 4 == 0 and self.Y % 100 != 0) or self.Y % 400 == 0:
            return [0, 31, 29, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
        else:
            return [0, 31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

    def valid_check(self):
        if not (1 <= self.M <= 12):
            return False
        
        M_list = self.get_max_days()
        
        if 1 <= self.D <= M_list[self.M] and self.Y > 0: 
            return True
        else:
            return False

    def next_day(self):
        if not self.valid_check():
             return "INVALID DATE"

        M_list = self.get_max_days()
        
        if self.D < M_list[self.M]:
            self.D += 1
        else:
            self.D = 1
            if self.M < 12:
                self.M += 1
            else:
                self.M = 1
                self.Y += 1
        
        return f"{self.D}/{self.M}/{self.Y}"

D, M, Y = map(int, input("Nhập ngày/tháng/năm: ").split("/"))
a = Date(D, M, Y)
print(a.valid_check())
if a.valid_check():
    print(a.next_day())
```

```python
#W8A4
class Calculator:
    def __init__(self, a: int, b:int):
        self.a = a
        self.b = b

    def add(self):
        return self.a + self.b

    def subtract(self):
        return self.a - self.b

    def product(self):
        return self.a * self.b
    
    def divide(self):
        if self.b == 0:
            return "Nhap so khac khong"
        else:
            return self.a / self.b
    
    def power(self):
        return self.a ** self.b
    
    def mod (self):
        if self.b == 0:
            return "Nhap so khac khong"
        else:
            return self.a % self.b
    
    def set_numbers (self, a, b):
        self.a = a
        self.b = b
    
def main():
        while True:
            try:
                a, b = map(float, input().split())
                break
            except ValueError:
                print("Nhap lai so")
        
        value = Calculator (a, b)

        while True:
            print("1.Cong")
            print("2.Tru")
            print("3.Nhan")
            print("4.Chia")
            print("5.Luy thua")
            print("6.Chia du")
            print("7.Nhap so khac")
        
            choice = input("Nhap chuc nang 1 - 7")

            if choice == '1':
                 result = value.add()
            elif choice == '2':
                result = value.subtract()
            elif choice == '3':
                result = value.product()
            elif choice == '4':
                result = value.divide()
            elif choice == '5':
                result = value.power()
            elif choice == '6':
                result = value.mod()
            elif choice == '7':
                try:
                    new_a = float(input("Nhap so: "))
                    new_b = float(input("Nhap so: "))
                    value.set_numbers(new_a, new_b)
                    continue
                except ValueError:
                    print("NHap lai")
                    continue
            else:
                print("Nhap lai")
                continue

            if choice in ['1', '2', '3', '4', '5', '6']:
                print(f"{result}")

            ask_exit = input("\nBạn có muốn thoát? (yes/no): ")

            if ask_exit in ["yes", "YES", "Yes"]:
                break

if __name__ == "__main__":
    main()
```

```python
class ShoppingCart:
    def __init__(self):
        self.products = []

    def add_item(self, name, price):

        item = {"name": name, "price": price} 
        self.products.append(item)

    def remove_item(self, name):
        found = False
        for i in self.products:
            if i["name"] == name:  
                self.products.remove(i)
                found = True
                break 

    def empty_check(self):
        if len(self.products) == 0:
            print("Giỏ hàng Trống")
        else:
            print("Giỏ hàng KHÔNG trống")

    def total_bill(self):
        total = 0
        for i in self.products:
            total += i["price"]
        return total

    def display_cart(self):
        for item in self.products:
            print(item)

    def remove_all(self):
        self.products.clear()

def main():
    cart = ShoppingCart()

    while True:
        print("2. Thêm sản phẩm")
        print("3. Xóa sản phẩm theo tên")
        print("4. Kiểm tra giỏ hàng rỗng")
        print("5. Tính tổng tiền")
        print("6. Hiển thị giỏ hàng")
        print("7. Xóa toàn bộ giỏ hàng")

        choice = input("Mời bạn nhập lựa chọn (2-7): ")

        if choice == '2':
            name = input("Nhập tên sản phẩm: ")
            try:
                price = float(input("Nhập giá sản phẩm: "))
                cart.add_item(name, price)
            except ValueError:
                print("Lỗi: Giá phải là số.")
        
        elif choice == '3':
            name = input("Nhập tên sản phẩm cần xóa: ")
            cart.remove_item(name)
        
        elif choice == '4':
            cart.empty_check()
        
        elif choice == '5':
            total = cart.total_bill()
            print(f"{total}")
        
        elif choice == '6':
            cart.display_cart()
            
        elif choice == '7':
            cart.remove_all()

        exit_choice = input("\nCó muốn thoát khỏi chương trình không? (yes/no): ")
        if exit_choice.lower() == "yes":
            break

if __name__ == "__main__":
    main()
```

```python
#W8A7
class Fraction:
    def __init__(self, num, den):
        a, b = abs(num), abs(den)
        while b > 0:
            a, b = b, a % b
        ucln = a

        num //= ucln
        den //= ucln
        
        if den < 0:
            num = -num
            den = -den
            
        self.num = num 
        self.den = den 

    def add(self, other):
        new_num = self.num * other.den + self.den * other.num
        new_den = self.den * other.den
        return Fraction(new_num, new_den)

    def sub(self, other):
        new_num = self.num * other.den - self.den * other.num
        new_den = self.den * other.den
        return Fraction(new_num, new_den)

    def mul(self, other):
        new_num = self.num * other.num
        new_den = self.den * other.den
        return Fraction(new_num, new_den)

    def div(self, other):
        new_num = self.num * other.den
        new_den = self.den * other.num
        return Fraction(new_num, new_den)

    def __str__(self):
        return f"{self.num}/{self.den}"

line = input().split()
a = int(line[0])
b = int(line[1])
op = line[2]
c = int(line[3])
d = int(line[4])

f1 = Fraction(a, b)
f2 = Fraction(c, d)

if op == '+':
    result = f1.add(f2)
elif op == '-':
    result = f1.sub(f2)
elif op == '*':
    result = f1.mul(f2)
elif op == '/':
    result = f1.div(f2)

if result:
    print(result)
```
