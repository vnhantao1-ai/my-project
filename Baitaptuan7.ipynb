```python
#W7A1
def binary_search(arr: list, left: int, right: int, target: int):
    while left <= right:
        mid = (left + right)//2
        if arr[mid] == target:
            return mid
        elif arr[mid] > target:
            right -= 1
        elif arr[mid] < target:
            left += 1
    return -1
arr = list(map(int, input().split()))
left, right = map(int, input().split())
target = int(input())
print(binary_search(arr, left, right, target))
```

```python
#W7A3
"Bubble sort"
def bubble_sort(arr):
    n = len(arr)
    for i in range(n - 1):
        for j in range(n - 1 - i):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr
"Selection sort"
def selection_sort(arr):
    n = len(arr)
    for i in range(n - 1):
        min_idx = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr
arr = list(map(int, input().split()))
print(selection_sort(arr))
```

```python
#W7A4
def most_occurences(arr: list):
    d = {}
    max_count = 0
    figure = 0
    for i in (arr):
        if i not in d:
            d[i] = 1
        else:
            d[i] += 1
    for idx, key in enumerate (d):
        if d[key] > max_count:
            max_count = d[key]
            figure = key
    return figure
arr = list(map(int, input().split()))
print(most_occurences(arr))
```

```python
#W7A5
def sum_target(arr: list, x: int):
    d = {}
    total = 0
    for i in arr:
        number = x - i
        if number in d:
            total += d[number]
        if i in d:
            d[i] += 1
        else:
            d[i] = 1
    return total
arr = list(map(int, input().split()))
x = int(input())
print(sum_target(arr, x))
```

```python
#W7A7
import ast
def length_of_lis(nums: list):
    tails = [nums[0]]
    n = len(nums)
    for i in range (1, n):
        left = 0
        right = len(tails)
        if nums[i] > tails[-1]:
            tails.append(nums[i])
        else:
            while left < right:
                mid = (left+right)//2
                if tails[mid] >= nums[i]:
                    right = mid
                else:
                    left = mid +1
            tails[left] = nums[i]
    return len(tails)
nums = ast.literal_eval(input())
print(length_of_lis(nums))
```

```python
#W7A9
# Đếm từ bên phải sang 
# tạo 1 list đã sắp xếp, rồi xét vị trí của i được xét, vị trí của i chính là số mà nó lớn hơn
import ast
def binary_search_tree(arr: list):
    sorted_list = []
    result = []
    n = len(arr)
    for i in range (n-1, -1, -1):
        left = 0
        right = len(sorted_list)
        while left < right:
            mid = (left+right)//2
            if sorted_list[mid] > arr[i]:
                right = mid
            elif sorted_list[mid] < arr[i]:
                left = mid + 1
        sorted_list.insert(left, arr[i])
        result.append(left)
    result.sort(reverse = True)
    return result
arr = ast.literal_eval(input())
print(binary_search_tree(arr))
```

```python
#W7A11
def count_occur(string: str, key_word: str):
    count = 0
    a = len(string)
    b = len(key_word)
    for i in range(a-b+1):
        if string[i:i + b] == key_word:
            count += 1
    return count
string = input()
key_word = input()
print(count_occur(string, key_word))
```

```python
#W7A12
def arrange(a, b, c, d):
    k = {"A": a,"B": b,"C": c,"D": d}
    l = [""] * 4
    for i in k:
        count = 0
        for j in k:
            if k[i].lower() > k[j].lower():
                count += 1
        pos = count
        while l[pos] != "":
            pos += 1
        l[pos] = i
    return l
a = input()
b = input()
c = input()
d = input()
print(" ".join(arrange(a, b, c, d)))
```

```python
#W7A13
def perfect_sequence(arr: list):
    d = {}
    l = []
    for i in arr:
        if i in d:
            d[i] += 1
        else:
            d[i] = 1
    for i in d:
        if (i+1) in d:
            l.append(d[i] + d[i+1])
    if l != []:
        return max(l)
    else:
        return "Khong ton tai"
arr = list(map(int, input().split()))
print(perfect_sequence(arr))
```

```python
#w7A14
def find_7(arr: list, k: int):
    if len(arr) != k:
        return "Nhap lai"
    else:
        n = len(arr)
        l = []
        if 7 in arr:
            for index in range(n):
                if arr[index] == 7:
                    l.append(index)
            l.sort(reverse = True)
            return l
        else:
            return "Not Found"
k = int(input())
arr = list(map(int, input().split()))

result = find_7(arr, k)
if isinstance(result, list):
    print(*result)
else:
    print(result)
```
