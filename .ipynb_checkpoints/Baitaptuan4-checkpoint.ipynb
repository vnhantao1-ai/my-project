```python
'''iết chương trình Python sử dụng một trong 2 loại vòng lặp for hoặc while để 
tính tống các số nguyên từ 1 đến n. (n <= 1000) '''
n = int(input())
x = 0
for i in range (1, n+1):
    x += i
print(x)
```

```python
'''Iết chương trình nhập vào một số tới khi được một số nguyên dương thì thôi. 
Kiểm tra số này có phải là số nguyên tố hay không? '''
import math as m
x = 0
for _ in range (10000):
    x = int(input())
    if x > 0: 
        print(f"{x} la 1 so duong")
        break
    else: 
        print("Nhập 1 số dương")
flag = True
if x < 2:
    flag = False
else:
    for i in range (2, int(m.sqrt(x)) +1):
        if x % i == 0:
            flag = False
            break
if flag == True: print(f"{x} la 1 so nguyen to")
else: print(f"{x} khong la 1 so nguyen to")
```

```python
'''iết chương trình tính giai thừa của một số nguyên n cho trước (n!=1×2×…×n) 
với 0 < n < 100. '''
try:
    n = int(input())
    tich = 1
    if 0 < n < 100:
        for i in range (1, n+1):
            tich *= i
        print(tich)
    else:
        print("Nhap dung pham vi")
except ValueError:
    print("Nhap dung so")
```

```python
'''iết chương trình nhập vào một số nguyên n và in ra màn hình số chữ số (trừ dấu) 
của số đó. Không dùng kiểu dữ liệu string. '''
try:
    n = int(input())
    x = abs(n)
    so_chu_so = 0
    for _ in range (10000):
        if x > 0:
            x //= 10
            so_chu_so += 1
        else:
            break
    print(so_chu_so)
except ValueError:
    print("Nhap dung so")
```

```python
'''iết chương trình đếm số lượng ước số chẵn của số nguyên dương n với n < 10^6. '''
try:
    n = int(input())
    so_uoc = 1
    if n > 0:
        for i in range (1, n//2+1):
            if n % i == 0 and i % 2 == 0:
                so_uoc += 1
        print(so_uoc)
    else:
        print("Nhap so nguyen")
except ValueError:
    print("Nhap so nguyen duong vao")
```

```python
'''ột người có tài khoản tiết kiệm ở ngân hàng và gửi vào X đồng với lãi suất là 
0.7% mỗi tháng. Viết chương trình tính số tiền sau N tháng người ấy rút được (cả 
gốc và lãi, bỏ qua phần lẻ thập phân). '''
try:
    tien_gui = float(input())
    so_thang = int(input())
    if so_thang > 0 and tien_gui > 0:
        for i in range (1, so_thang+1):
            tien_gui *= 1.007
        print(tien_gui)
    else:
        print("Nhap so duong")
except ValueError:
    print("Nhap so dung")
```

```python
'''hập hai số từ bàn phím và kiểm tra xem chúng có phải là cặp số thân thiết không. 
Nếu có, in ra màn hình "true", ngược lại, in ra "false"'''
try:
    a, b = map(int, input().split())
    if a > 0 and b>0:
        tong_uoc_a = 0
        for i in range (1, a//2+1):
            if a % i == 0:
                tong_uoc_a += i
        if tong_uoc_a == b:
            print("true")
        else:
            print("false")
    else:
        print("Nhap 2 so lon hon 0")
except ValueError:
    print("Nhap gia tri dung")
```

```python
'''Cho hai số m và n là hai số nguyên dương nhập từ bàn phím. Viết chương trình 
tìm ước chung lớn nhất của hai số m và n. '''
try:
    m, n = map(int, input().split())
    if m > 0 and n > 0:
        if m > n:
            for i in range (10000):
                m, n = n, m%n
                if n == 0:
                    break
            print(m)
        else:
            for i in range (10000):
                n, m = m, n % m
                if m == 0:
                    break
            print(n)
    else:
        print("Nhap so duong")
except ValueError:
    print("Nhap lai")
```

```python
'''hập vào số nguyên dương n (n ≥ 2). Hãy in ra ước số nguyên tố lớn nhất của n." '''
try:
    n = int(input())
    if n >= 2:
        i = 2
        for _ in range (10000):   # Tim bằng cách chia dần cho đến khi ước bình lớn hơn n
            if i**2 > n:
                break
            if n % i == 0:
                n //= i
            else:
                i += 1
        print(n)
    else:
        print("Nhap lai")
except ValueError:
    print("Nhap lai")
```

```python
'''hập vào số nguyên dương n. In ra tất cả các số chính phương ≤ n có các chữ số 
đôi một khác nhau (in theo thứ tự tăng dần, cùng hàng, cách nhau bởi một dấu 
cách). Nếu không có, in ra (no number). '''
try:
    n = int(input())
    if n >= 0:
        for i in range (n+1):
            if i**2 <= n:
                print(i**2, end = " ")
            elif i**2 > n:
                break
    else:
        print("no number")
except ValueError:
    print("Nhap lai")
```
