```python
#W7A16
def perfect_arr(arr: list):
    d = {}
    for i in arr:
        if i in d:
            d[i] += 1
        else:
            d[i] = 1 

    remove = 0
    for i in d:
        for i in d:
            if d[i] < i:
                remove += d[i]
            elif d[i] > i:
                remove += d[i] - i
    if remove == 0:
        return "Yes"
    else:
        return f"No, remove {remove}"
    
arr = list(map(int, input().split()))
print(perfect_arr(arr))
```

```python
#W7A17
def boyer_moore(arr: list):
    candidate = None
    count = 0
    n = len(arr)
    
    for num in arr:
        if count == 0:
            candidate = num
            count = 1
        elif candidate != num:
            count -= 1
        elif candidate == num:
            count += 1

    real_count = 0

    for num in arr:
        if candidate == num:
            real_count += 1

    if real_count > (n/2):
        return candidate
    else:
        pass

arr = list(map(int, input().split()))
print(boyer_moore(arr))
```

```python
#W7A18
def arrange_arr(arr: list):
    even = []
    odd = []
    
    for i in arr:
        if i % 2 == 0:
            even.append(i)
        else:
            odd.append(i)
            
    even.sort()
    odd.sort(reverse = True)
    a = even + odd
    return a
arr = list(map(int, input().split()))
print(*arrange_arr(arr))
```

```python
#W7A19
def lite_merge_sort(arr1: list, arr2: list):
    n1, n2 = len(arr1), len(arr2)
    i = j =0
    l = []
    
    if arr1[0] < arr1[-1] and n1 >= 2:
        while i < n1 and j < n2:
            if arr1[i] <= arr2[j]:
                l.append(arr1[i])
                i += 1
            elif arr1[i] > arr2[j]:
                l.append(arr2[j])
                j += 1
        l.extend(arr1[i:])
        l.extend(arr2[j:])
        return l
        
    elif arr1[0] > arr1[-1] and n1 >= 2:
        while i < n1 and j < n2:
            if arr1[i] >= arr2[j]:
                l.append(arr1[i])
                i += 1
            elif arr1[i] < arr2[j]:
                l.append(arr2[j])
                j += 1
        l.extend(arr1[i:])
        l.extend(arr2[j:])
        return l
        
arr1 = list(map(int, input().split()))
arr2 = list(map(int, input().split()))
print(lite_merge_sort(arr1, arr2))
```

```python
#W7A21
def total_days(N: int, T: int):
    total = N
    max_day = [0, 31, 29, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

    if T < 1 or T > 12 or N < max_day[T] or N > max_day[T]:
        return "invalid"
    
    for i in range(T):
        total += max_day[i]

    return total
N, T = map(int, input().split())
print(total_days(N, T))
```

```python
#W7A22
def most_appearance(arr: list):
    l = []
    d = {}
    n = len(arr)
    alphabet_vn = [
    'a', 'ă', 'â', 'b', 'c', 'd', 'đ', 'e', 'ê', 'g', 'h', 'i', 
    'k', 'l', 'm', 'n', 'o', 'ô', 'ơ', 'p', 'q', 'r', 's', 't', 
    'u', 'ư', 'v', 'x', 'y']
    
    for i in range (n):
        arr[i] = arr[i].split()
        l.append(arr[i][-1])

    for name in l:
        if name not in d:
            d[name] = 1
        else:
            d[name] += 1

    count = 0
    proper_name = ""
    
    for name in d:
        if d[name] > count:
            count = d[name]
            proper_name = name
        elif d[name] == count:
            if alphabet_vn.index(name[0].lower()) < alphabet_vn.index(proper_name[0].lower()):
                proper_name = name

    return proper_name

n = int(input())
arr = []

for i in range(n):
    a = input()
    arr.append(a)

print(most_appearance(arr))
```

```python
#W7A24
def remove_word(string: str):
    a = string.split()
    n = len(a)
    new_l = []

    for i in range (n-1):
        if a[i] != a[i+1]:
            new_l.append(a[i])
    new_l.extend(a[i+1:])
    b = " ".join(new_l)

    return b

string = input()
print(remove_word(string))
```

```python
#W7A25
def inves_max(string: str, k: int):
    new_l = []
    number_list = [int(i) for i in string]
    remove_count = len(string) - k

    for i in number_list:
        while remove_count > 0 and len(new_l) > 0 and new_l[-1] < i:
            new_l.pop()
            remove_count -= 1
        new_l.append(i)

    result = "".join(map(str, new_l))

    return result
string = input()
k = int(input())

print(inves_max(string, k))
```

```python
#W7A26
def split_number_from_string(string: str):
    numbers = [] 
    current_num = "" 
    string += " " 
    
    for char in string:
        if char.isdigit():
            current_num += char
        else:
            if current_num != "":
                numbers.append(int(current_num)) 
                current_num = ""

    if len(numbers) == 0:
        return 0, 0.0 
        
    total = sum(numbers)
    average = total / len(numbers)
    
    return total, average

input_str = input("Nhập chuỗi: ")
total, avg = split_number_from_string(input_str)
print(f"Tổng: {total}, Trung bình: {avg}")
```

```python
#W7A27
def check_bracket(string: str):
    stack = []
    bracket_map = {")": "(", "]": "[", "}": "{"} 

    for char in string:
        if char in ["(", "[", "{"]:
            stack.append(char)
        
        elif char in [")", "]", "}"]:
            if stack == [] or stack[-1] != bracket_map[char]:
                return "No"
            stack.pop()
        
    if len(stack) == []:
        return "Yes"
    else:
        return "No"

string = input("Nhập chuỗi ngoặc: ")
print(check_bracket(string))
```

```python
#W7A30
def find_sequence(arr: list):
    new_l = []

    for i in arr:
        if new_l == [] or i >= new_l[-1]:
            new_l.append(i)
            
        else:
            left = 0
            right = len(new_l)
            while left < right:
                mid = (left+right) // 2
                if new_l[mid] >= i:
                    right = mid
                else:
                    left = mid + 1
            new_l[left] = i

    return new_l
    
arr = list(map(int, input().split()))
print(find_sequence(arr))
```

```python
#W7A31
def swap_pos_to_min(arr: list):
    n = len(arr)
    pos_of_min = [0]*n
    pos_of_min[n-1] = n-1

    for i in range (n - 2, -1, -1):
        if arr[i] < arr[pos_of_min[i+1]]:
            pos_of_min[i] = i
        elif arr[i] == arr[pos_of_min[i+1]]:
            pos_of_min[i] = i + 1
        else:
            pos_of_min[i] = pos_of_min[i+1]

    swap = False
    for i in range (len(pos_of_min)):
        if arr[i] > arr[pos_of_min[i]]:
            arr[i], arr[pos_of_min[i]] = arr[pos_of_min[i]], arr[i]
            swap = True
            break

    if not swap:
        arr[n-1], arr[n-2] = arr[n-2], arr[n-1]

    result = "".join(map(str, arr))

    return result

arr = list(map(int, input()))
print(swap_pos_to_min(arr))
```
