```python
#W7A16
def perfect_arr(arr: list):
    d = {}
    for i in arr:
        if i in d:
            d[i] += 1
        else:
            d[i] = 1 

    remove = 0
    for i in d:
        for i in d:
            if d[i] < i:
                remove += d[i]
            elif d[i] > i:
                remove += d[i] - i
    if remove == 0:
        return "Yes"
    else:
        return f"No, remove {remove}"
    
arr = list(map(int, input().split()))
print(perfect_arr(arr))
```

```python
#W7A17
def boyer_moore(arr: list):
    candidate = None
    count = 0
    n = len(arr)
    
    for num in arr:
        if count == 0:
            candidate = num
            count = 1
        elif candidate != num:
            count -= 1
        elif candidate == num:
            count += 1

    real_count = 0

    for num in arr:
        if candidate == num:
            real_count += 1

    if real_count > (n/2):
        return candidate
    else:
        pass

arr = list(map(int, input().split()))
print(boyer_moore(arr))
```

```python
#W7A18
def arrange_arr(arr: list):
    even = []
    odd = []
    
    for i in arr:
        if i % 2 == 0:
            even.append(i)
        else:
            odd.append(i)
            
    even.sort()
    odd.sort(reverse = True)
    a = even + odd
    return a
arr = list(map(int, input().split()))
print(*arrange_arr(arr))
```

```python
#W7A19
def lite_merge_sort(arr1: list, arr2: list):
    n1, n2 = len(arr1), len(arr2)
    i = j =0
    l = []
    
    if arr1[0] < arr1[-1] and n1 >= 2:
        while i < n1 and j < n2:
            if arr1[i] <= arr2[j]:
                l.append(arr1[i])
                i += 1
            elif arr1[i] > arr2[j]:
                l.append(arr2[j])
                j += 1
        l.extend(arr1[i:])
        l.extend(arr2[j:])
        return l
        
    elif arr1[0] > arr1[-1] and n1 >= 2:
        while i < n1 and j < n2:
            if arr1[i] >= arr2[j]:
                l.append(arr1[i])
                i += 1
            elif arr1[i] < arr2[j]:
                l.append(arr2[j])
                j += 1
        l.extend(arr1[i:])
        l.extend(arr2[j:])
        return l
        
arr1 = list(map(int, input().split()))
arr2 = list(map(int, input().split()))
print(lite_merge_sort(arr1, arr2))
```

```python
#W7A20
def total(arr1: list, k: int):
    l = []
    for i in arr1:
        if len(str(i)) == k:
            l.append(i)
    return sum(l)
arr = list(map(int, input().split()))
k = int(input())
print(total(arr, k))
```

```python
#W7A21
def total_days(N: int, T: int):
    total = N
    max_day = [0, 31, 29, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

    if T < 1 or T > 12 or N < max_day[T] or N > max_day[T]:
        return "invalid"
    
    for i in range(T):
        total += max_day[i]

    return total
N, T = map(int, input().split())
print(total_days(N, T))
```

```python
#W7A22
def most_appearance(arr: list):
    l = []
    d = {}
    n = len(arr)
    alphabet_vn = [
    'a', 'ă', 'â', 'b', 'c', 'd', 'đ', 'e', 'ê', 'g', 'h', 'i', 
    'k', 'l', 'm', 'n', 'o', 'ô', 'ơ', 'p', 'q', 'r', 's', 't', 
    'u', 'ư', 'v', 'x', 'y']
    
    for i in range (n):
        arr[i] = arr[i].split()
        l.append(arr[i][-1])

    for name in l:
        if name not in d:
            d[name] = 1
        else:
            d[name] += 1

    count = 0
    proper_name = ""
    
    for name in d:
        if d[name] > count:
            count = d[name]
            proper_name = name
        elif d[name] == count:
            if alphabet_vn.index(name[0].lower()) < alphabet_vn.index(proper_name[0].lower()):
                proper_name = name

    return proper_name

n = int(input())
arr = []

for i in range(n):
    a = input()
    arr.append(a)

print(most_appearance(arr))
```

```python
#W7A23
def count_name(string: str):
    l = string.split()
    count = 0

    for word in l:
        if word[0] == word[0].upper():
            count += 1

    return count

string = input()
print(count_name(string))
```

```python
#W7A24
def remove_word(string: str):
    a = string.split()
    n = len(a)
    new_l = []

    for i in range (n-1):
        if a[i] != a[i+1]:
            new_l.append(a[i])
    new_l.extend(a[i+1:])
    b = " ".join(new_l)

    return b

string = input()
print(remove_word(string))
```
